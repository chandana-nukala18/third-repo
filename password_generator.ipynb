Basic password generator

```python
import random
import string

def generate_password(length,use_uppercse,use_lowercase,use_digits,use_specials):
  character_pool = ''
  if use_uppercse:
    character_pool += string.ascii_uppercase
  if use_lowercase:
    character_pool += string.ascii_lowercase
  if use_digits:
    character_pool += string.digits
  if use_specials:
    character_pool += string.punctuation

  if not character_pool:
    raise ValueError("At least one character set must be selected.")

  password = ''.join(random.choice(character_pool) for _ in range(length))
  return password

def get_user_input():
  try:
    length = int(input("Enter the password length:"))
    if length <= 0:
      raise ValueError("password length must be greater than zero.")
  except ValueError as e:
    print(f"Invalid input: {e}")
    return

  use_uppercase = input("Include uppercase letters?(y/n:)").strip().lower() == 'y'
  use_lowercase = input("Include lowercase letters?(y/n:)").strip().lower() == 'y'
  use_digits = input("Include digits?(y/n:)").strip().lower() == 'y'
  use_specials = input("Include special characters?(y/n:)").strip().lower() == 'y'

  try:
      password = generate_password(length, use_uppercase, use_lowercase, use_digits, use_specials)
      print(f"\nGenerated password: {password}")
  except ValueError as e:
      print(f"Error: {e}")

if __name__ == "__main__":
  get_user_input()
```

advanced password genrator


```python
import random
import string
from IPython.display import display, HTML, Javascript
import ipywidgets as widgets


#widgets for input
length_slider = widgets.IntSlider(value=12, min=4, max=64, description='Length:')
upper_checkbox = widgets.Checkbox(value=True, description='uppercase Letters')
lower_checkbox = widgets.Checkbox(value=True, description='Lowercase Letters')
digits_checkbox = widgets.Checkbox(value=True, description='Digits')
symbols_checkbox = widgets.Checkbox(value=True,description='Symbols')
generate_button = widgets.Button(description='Generate password')
copy_button = widgets.Button(description='copy to clipboard', button_style='success')
output = widgets.Output()

password_display = widgets.Text(
    value='',
    description='password:',
    layout=widgets.Layout(width='100%'),
    style={'description_width':'initial'}
    )


#password generation logic
def generate_password(_):
    length = length_slider.value
    use_upper = upper_checkbox.value
    use_lower = lower_checkbox.value
    use_digits = digits_checkbox.value
    use_symbols = symbols_checkbox.value

    character_pool = ''
    selected_types = []


    if use_upper:
        character_pool += string.ascii_uppercase
        selected_types.append(string.ascii_uppercase)
    if use_lower:
        character_pool += string.ascii_lowercase
        selected_types.append(string.ascii_lowercase)
    if use_digits:
        character_pool += string.digits
        selected_types.append(string.digits)
    if use_symbols:
        character_pool += string.punctuation
        selected_types.append(string.punctuation)
    if not character_pool:
        password_display.value = "Error: select at least one character type!"
        return

    if length < len(selected_types):
        password_display.value = f"Error: Length must be at least {len(selected_types)}"
        return
     #Ensure inclusion of atleast one from each selected category
    password = [random.choice(char_set) for char_set in selected_types]
    password += random.choices(character_pool, k=length - len(password))
    random.shuffle(password)
    password_display.value = ''.join(password)


#javascript clipboard integration
def copy_to_clipboard(_):
    display(Javascript("""
         const text = document.querySelector('input[type="text"]').value;
         navigator.clipboard.writeText(text).then(() => {
            alert("password copied to C vlipboard!");
         });
      """))


generate_button.on_click(generate_password)
copy_button.on_click(copy_to_clipboard)

#layout
ui = widgets.VBox([
    length_slider,
    upper_checkbox, lower_checkbox, digits_checkbox, symbols_checkbox,
    widgets.HBox([generate_button, copy_button]),
    password_display,
    output
])

display(ui)

```
